## PDF tables extraction process.

The following is the process used for the data extraction, it will vary from pdf to pdf even if they seem to have the same structure but it is fairly similar, for this reason i will only explain the first one of the year 2018.

### 2018

In this first step the data is loaded from the pdf by a libary called tabula which is handled in a separate script in the file data_extraction.py.  
The paramteters needed are the path of the file and the range of pages the table is.

```python
general_data = de.pdf_table_extraction("./DataPDFs/dt79.pdf","6-7")[0]
general_data.head()
```

Since the data collection from the library extracts all the rows in each page the dataframe has an entry with the names of the columns and it is eliminated int the next cell by obtaining all the rows excepting the one with the column names.

```python
general_data = general_data[general_data['Entidad'] != 'Entidad']
general_data
```

For the next step a second table will be extracted from the pdf  containig information about the traffic incidents.

```python

transit_accidents = de.pdf_table_extraction("./DataPDFs/dt79.pdf","9-11")[0]
transit_accidents.head()
```

As shown above the collected data has to be poceesed do correct the mistakes made by the library by correcting the names and dropping NaN columns.  
Also because of the different names for the same atribute ('Entidad') the name is changed the with the same one of the first table for future mergeing.

```python
columns = transit_accidents.columns
for column in columns:
    if pd.notna(transit_accidents[column][0]):
        transit_accidents.rename(columns={column: transit_accidents[column][0]}, inplace=True)
transit_accidents = transit_accidents.loc[:, ~transit_accidents.columns.str.contains("Unnamed")]
transit_accidents = transit_accidents.dropna()
transit_accidents = de.column_cleaning([transit_accidents])[0]
transit_accidents.rename(columns={'Entidad Federativa': 'Entidad'}, inplace=True)
len(transit_accidents.columns)
```

Here the mergeing is handled as a SQL merge using the pandad merge function and the column year is added for future diferenciation when added to a bigger dataset of later years.

```python
traffic_df_2018 = pd.merge(general_data, transit_accidents, on='Entidad', how='inner')
traffic_df_2018['Year'] = 2018
traffic_df_2018
```

#### Standardize columns

In this section the columns are standardized for future merging with the other data extracted utilizing the standard names created in the beggining.

```python
def standardize_columns(standard_columns, df):
    if len(df.columns) == len(standard_columns):
        df.columns = standard_columns
        return df
    else:
        return -1
    
```

```python
traffic_df_2018 = standardize_columns(standard_columns, traffic_df_2018)
traffic_df_2018.head()
```

Now that tables that will be used for the analysis are correctly extracted and merged the same process will be done to all the PDF's each one corresponding to a year.

### 2019

```python
general_data = de.pdf_table_extraction("./DataPDFs/dt82.pdf","5-6")[0]
general_data = general_data[general_data['Entidad'] != 'Entidad']

transit_accidents = de.pdf_table_extraction("./DataPDFs/dt82.pdf", "8")[0]
transit_accidents = transit_accidents.drop(columns=['Unnamed: 2'])
transit_accidents.columns = correct_column_names
transit_accidents.drop(index=0, inplace=True)

temp_table = de.pdf_table_extraction("./DataPDFs/dt82.pdf", "9")[0]
temp_table = temp_table.drop(columns=['Unnamed: 2', 'Unnamed: 4'])
temp_table.columns = correct_column_names
temp_table.drop(index=0, inplace=True)

transit_accidents = pd.concat([transit_accidents, temp_table], ignore_index=True)

temp_table = de.pdf_table_extraction("./DataPDFs/dt82.pdf", "10")[0]
temp_table = temp_table.drop(columns=['Unnamed: 2'])
temp_table.columns = correct_column_names
temp_table.drop(index=0, inplace=True)

transit_accidents = pd.concat([transit_accidents, temp_table], ignore_index=True)

traffic_df_2019 = pd.merge(general_data, transit_accidents, on='Entidad', how='inner')
traffic_df_2019['Year'] = 2019
traffic_df_2019 = standardize_columns(standard_columns, traffic_df_2019)
traffic_df_2019.head()
```

### 2020

```python
general_data = de.pdf_table_extraction("./DataPDFs/dt84.pdf","5-6")[0]
general_data = general_data.drop(columns=['Unnamed: 0'])
general_data = general_data.dropna()
general_data = general_data[general_data['Entidad'] != 'Entidad']

transit_accidents = de.pdf_table_extraction("./DataPDFs/dt84.pdf","8-9")[0]
transit_accidents = transit_accidents.drop(columns=['Unnamed: 0'])
transit_accidents.drop(index=[0,1], inplace=True)
transit_accidents.dropna(inplace=True)
transit_accidents.columns = correct_column_names

traffic_df_2020 = pd.merge(general_data, transit_accidents, on='Entidad', how='inner')
traffic_df_2020['Year'] = 2020
traffic_df_2020 = standardize_columns(standard_columns, traffic_df_2020)
traffic_df_2020.head()
```

#### 2021

```python
general_data = de.pdf_table_extraction("./DataPDFs/dt86.pdf","5-6")[0]
general_data = general_data.drop(columns=['Unnamed: 0'])
general_data = general_data.dropna()
general_data = general_data[general_data['Entidad'] != 'Entidad']

transit_accidents = de.pdf_table_extraction("./DataPDFs/dt86.pdf","8-9")[0]
transit_accidents = transit_accidents.drop(columns=['Unnamed: 0'])
transit_accidents.drop(index=[0,1], inplace=True)
transit_accidents.dropna(inplace=True)
transit_accidents.columns = correct_column_names

traffic_df_2021 = pd.merge(general_data, transit_accidents, on='Entidad', how='inner')
traffic_df_2021['Year'] = 2021
traffic_df_2021 = standardize_columns(standard_columns, traffic_df_2021)
traffic_df_2021.head()
```

```python
general_data = de.pdf_table_extraction("./DataPDFs/dt91.pdf","5-6")[0]
general_data = general_data.drop(columns=['Unnamed: 0'])
general_data = general_data.dropna()
general_data = general_data[general_data['Entidad'] != 'Entidad']
general_data

transit_accidents = de.pdf_table_extraction("./DataPDFs/dt91.pdf","7-8")[0]
transit_accidents = transit_accidents.drop(columns=['Unnamed: 0'])
transit_accidents.drop(index=[0,1], inplace=True)
transit_accidents.dropna(inplace=True)
transit_accidents.columns = correct_column_names
transit_accidents

traffic_df_2022 = pd.merge(general_data, transit_accidents, on='Entidad', how='inner')
traffic_df_2022['Year'] = 2022
traffic_df_2022 = standardize_columns(standard_columns, traffic_df_2022)
traffic_df_2022.head()
```

With all of the data extracted all the tables are concatenated in a final dataframe.

```python
final_df = pd.concat([traffic_df_2018, traffic_df_2019], ignore_index=True)
final_df = pd.concat([final_df, traffic_df_2020], ignore_index=True)
final_df = pd.concat([final_df, traffic_df_2021], ignore_index=True)
final_df = pd.concat([final_df, traffic_df_2022], ignore_index=True)

final_df
```

Finally the extracted dataset is saved into a csv to be used in any analysis or modeling.

```python
final_df.to_csv("./data/accidentes.csv", index=False)
```
